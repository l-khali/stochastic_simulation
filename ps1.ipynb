```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
rng = np.random.default_rng(42)
```

```python
samples = []
for sim in range(1000000):
    x = sum(rng.uniform(0,1,12)) - 6
    samples.append(x)
```

```python
plt.hist(samples)
```

```python
x = np.linspace(-6,6,10000)
plt.plot(x, ss.norm.pdf(x))
```

```python
samples2 = []

for sim in range(100000):
    samples2.append(-np.log(rng.uniform(0,1)))
```

```python
plt.hist(samples2)
```
